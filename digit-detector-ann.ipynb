# Import Libraries

```python
import numpy as np # linear algebra
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from keras.utils.np_utils import to_categorical # convert to one-hot-encoding
from keras.models import Sequential
```

```python
df = pd.read_csv("/kaggle/input/digit-recognizer/train.csv")
df.head()
```

```python
# Checking for the missing values
df.isnull().any().describe()
```

```python
# the label row :
X = df.iloc[:, 1:].values
# the pixel entries
Y = df.iloc[:, 0].values
```

```python
# Split into train and test set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 572)
```

```python
# One hot coding the labels
from keras.utils.np_utils import to_categorical
y_train, y_test = map(to_categorical,(y_train,y_test))
```

**Standardisation**

In this case, in order for gradient descent to converge faster we divide all the pixel datas by 255 to get a value between 0-1 for each pixel.


```python
X_train, X_test = X_train/255 , X_test/255
```

```python
fig, ax = plt.subplots(2, 1, figsize=(16,8))
ax[0].plot(X_train[5]) 
ax[0].set_title('784x1 data')
ax[1].imshow(X_train[5].reshape(28,28), cmap='gray')
ax[1].set_title('Corresponding 28 x 28 image')
plt.tight_layout()
```

```python
X_train.shape
```

# Model building

```python
from tensorflow import keras

NN = keras.models.Sequential()
NN.add(keras.Input(shape=(784,)))
NN.add(keras.layers.Dense(10, activation = 'relu'))
NN.add(keras.layers.Dense(10, activation = 'softmax')) # last layer
```

```python
NN.summary()
```

```python
NN.compile(optimizer = 'sgd',
           loss ='categorical_crossentropy' ,
           metrics = ['accuracy'] )
```

```python
X_train
```

```python
y_train
```

```python
NN.fit(X_train, y_train, batch_size=100, epochs=20)
```

```python
NN.evaluate(X_test, y_test)
```
